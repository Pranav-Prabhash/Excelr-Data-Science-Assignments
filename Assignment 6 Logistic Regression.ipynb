##   Problem Statement 
Input variables:
    bank client data
     
Output variable -> y
y -> Whether the client has subscribed a term deposit or not 
Binomial ("yes" or "no")

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#Reading dataset
data= pd.read_csv('bank1.csv',sep=';')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# One-Hot Encoding of categorical variables
data1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
#to see all coloumns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1 , 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

## Model Building

```python
# Dividing our data into input and output variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

## Model Predictions

```python
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y,'predicted_y':y_pred})
y_pred_df
```

## Testing Modal Accuracy

```python
#Confusion Matrix
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39156+1162)/(39156+766+4127+1162)
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

```python
#Classification report
## Precision – What percent of your predictions were correct?
## Recall – What percent of the positive cases did you catch? 
## F1 score – What percent of positive predictions were correct?

print(classification_report(y,y_pred))
```

```python
# intercept value
classifier.intercept_
```

```python
# Other coeffieicents
classifier.coef_
```

```python
# ROC Curve plotting and finding AUC value
fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba (x)[:,1])

auc = roc_auc_score(y, y_pred)


plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
print('auc accuracy:',auc)
```
