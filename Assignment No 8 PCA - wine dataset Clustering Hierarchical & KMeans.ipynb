## Data Preprocessing

```python
#Normalising data

#importing scale Library
from sklearn.preprocessing import scale

#Changing to array 
wine1.arr = wine1.values

#Applying normal scale
wine_norm = scale(wine1)
wine_norm
```

##  PCA

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#Variance Plot for PCA


plt.figure(figsize=(10,5))
plt.plot(var1,color="red")
plt.title('SCREE PLOT')
plt.xlabel("PCA")
plt.ylabel("Cumulative Variance")
plt.grid()
plt.show()
```

Observation 
* Around 75% of the variance of the data is captured by the first 3 PCAs

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,10))
sns.scatterplot(data=final_df)
```

## Checking with other Clustering Algorithms

### Hierarchical Clustering

```python
#Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
lk=sch.linkage(wine_norm,'complete')
dendrogram=sch.dendrogram(lk)
```

```python
# Create Clusters (y)
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc_cluster = hc.fit_predict(wine_norm)
```

```python
set(hc_cluster)
```

```python
# Concating Labels with main dataset copy
wine3 =wine.copy()
wine3["Hierarchical Labels"] =hc_cluster
wine3
```

```python
wine3['Hierarchical Labels'].value_counts()
```

## KMeans Clustering

```python
# Import Libraries
from sklearn.cluster import KMeans
```

```python
# elbow graph
wcss=[]

for i in range (1,13):
    km=KMeans(n_clusters=i)
    km.fit_predict(wine_norm)
    wcss.append(km.inertia_)
    
plt.figure(figsize=(8,5))    
plt.plot(range(1,13),wcss)
plt.grid()
plt.show()
```

Observation 
* No of clusters = 3 is the elbow point

### Build Cluster algorithm using K=3

```python
# Cluster algorithm using K=3
kcluster=KMeans(3,random_state=30).fit(wine_norm)
kcluster
```

```python
kcluster.labels_
```

```python
set(kcluster.labels_)
```

```python
# Assign clusters to the data set
wine4=wine.copy()
wine4['k_clusterid']=kcluster.labels_
wine4
```
